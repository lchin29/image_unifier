```python
import colorsys as cs
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image,ImageEnhance
```

```python
image = Image.open('cat_test.jpeg')
image
```

```python
rgb_image = np.asarray(image)
hls_image = [[ list(cs.rgb_to_hls(*rgb)) for rgb in row] for row in rgb_image]
hls_image
```

```python
brightened = [[ hls for hls in row] for row in hls_image]
rgb_bright = [[ np.round(np.asarray(cs.hls_to_rgb(*rgb))).astype(int) for rgb in row] for row in brightened]


# b_image = Image.fromarray(rgb_bright,"RGB")
plt.imshow(rgb_bright)
```

```python
print(rgb_image[224][224])
print(hls_image[224][224])
x = cs.rgb_to_hls(*rgb_image[224][224])
print(x)
print(cs.hls_to_rgb(*x))
print(cs.hls_to_rgb(*brightened[224][224]))
```

```python
data = np.asarray(image)
factor = .4
enhancer = ImageEnhance.Brightness(image)
output = enhancer.enhance(factor)
output
```
